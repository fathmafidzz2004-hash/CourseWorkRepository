```python
import pandas as pd

customers = pd.read_csv("CourseWorkRepository/customers.csv")
orders = pd.read_csv("CourseWorkRepository/orders.csv")
drivers = pd.read_csv("CourseWorkRepository/drivers.csv")
vehicles = pd.read_csv("CourseWorkRepository/vehicles.csv")
complaints = pd.read_csv("CourseWorkRepository/complaints.csv")
incidents = pd.read_csv("CourseWorkRepository/incidents.csv")
hubs = pd.read_csv("CourseWorkRepository/hubs.csv")
app_events = pd.read_csv("CourseWorkRepository/app_events.csv")
```

```python
for name, df in {
    "customers": customers,
    "orders": orders,
    "drivers": drivers,
    "vehicles": vehicles,
    "complaints": complaints,
    "incidents": incidents,
    "hubs": hubs,
    "app_events": app_events
}.items():
    print("\n", name)
    print(df.head())
    print(df.info())
```

```python


customers["home_zone"] = customers["home_zone"].str.upper()
orders["pickup_zone"] = orders["pickup_zone"].str.upper()
orders["dropoff_zone"] = orders["dropoff_zone"].str.upper()
drivers["base_zone"] = drivers["base_zone"].str.upper()
vehicles["assigned_zone"] = vehicles["assigned_zone"].str.upper()
app_events["zone_context"] = app_events["zone_context"].str.upper()



customers["preferred_channel"] = customers["preferred_channel"].fillna("Unknown")
customers["loyalty_score"] = customers["loyalty_score"].fillna(customers["loyalty_score"].mean())

drivers["training_score"] = drivers["training_score"].fillna(drivers["training_score"].mean())

vehicles["battery_health_pct"] = vehicles["battery_health_pct"].fillna(
    vehicles["battery_health_pct"].mean()
)

complaints["compensation_amount"] = complaints["compensation_amount"].fillna(0)

orders["booking_channel"] = orders["booking_channel"].fillna("Unknown")
```

```python
import sqlite3
```

```python
conn = sqlite3.connect("northstar.db")
```

```python
customers.to_sql("customers", conn, if_exists="replace", index=False)
orders.to_sql("orders", conn, if_exists="replace", index=False)
drivers.to_sql("drivers", conn, if_exists="replace", index=False)
complaints.to_sql("complaints", conn, if_exists="replace", index=False)
```

```python
query = """
SELECT service_type, COUNT(order_id) AS total_orders
FROM orders
GROUP BY service_type;
"""

pd.read_sql(query, conn)
```

```python
query = """
SELECT employment_type, AVG(driver_rating) AS avg_rating
FROM drivers
GROUP BY employment_type;
"""

pd.read_sql(query, conn)
```

```python
query = """
SELECT c.customer_type, AVG(o.order_value) AS avg_order_value
FROM customers c
JOIN orders o
ON c.customer_id = o.customer_id
GROUP BY c.customer_type;
"""

pd.read_sql(query, conn)
```
